```python
import pandas as pd
```

```python
df = pd.read_csv('daily_sales_data_0.csv')
```

```python
df.head()
```

```python
df = df[df['product'] == 'pink morsel']
df
```

```python
df['price'] = df['price'].str.replace('$', '').astype(float)
df.info()
```

```python
df['sales'] = df['price'] * df['quantity']
df.drop(columns=['price', 'quantity'], inplace=True)

df = df[['product', 'sales', 'date', 'region']]
df
```

```python
df_1 = pd.read_csv('daily_sales_data_1.csv')
df_1.head()
```

```python
df_1 = df_1[df_1['product'] == 'pink morsel']
df_1['price'] = df_1['price'].str.replace('$', '').astype(float)
df_1['sales'] = df_1['price'] * df_1['quantity']
df_1.drop(columns=['price', 'quantity'], inplace=True)
df_1 = df_1[['product', 'sales', 'date', 'region']]
df_1
```

```python
df_2 = pd.read_csv('daily_sales_data_2.csv')
df_2.head()
```

```python
df_2 = df_2[df_2['product'] == 'pink morsel']
df_2['price'] = df_2['price'].str.replace('$', '').astype(float)
df_2['sales'] = df_2['price'] * df_2['quantity']
df_2.drop(columns=['price', 'quantity'], inplace=True)
df_2 = df_2[['product', 'sales', 'date', 'region']]
df_2
```

```python
# Merge df, df_1, df_2
data = pd.concat([df, df_1, df_2])
data['date'] = pd.to_datetime(data['date']) # Convert date to datetime (optional)
data
```

```python
data.to_csv('pink_morsel_sales.csv', index=False)
```
